```python
import pandas as pd
import numpy as np
```

```python
#import the original security wait times data for all U.S. airports
waittimes_df = pd.read_excel("security_wait_times.xls")
waittimes_df
```

```python
#deal with merged cells of one airport corresponding to multiple checkpoints from the original excel file
waittimes_df['Code'] = pd.Series(waittimes_df['Code']).fillna(method='ffill')
waittimes_df['Airport'] = pd.Series(waittimes_df['Airport']).fillna(method='ffill')
```

```python
#calculate average security wait times of each airport based on data available
#store the values in a new column
avg_wait_time = waittimes_df.groupby("Code")[waittimes_df.columns[3:]].apply(np.nanmean)
waittimes_df['avg_wait_time'] = avg_wait_time[waittimes_df.Code].values
```

```python
#remname column names so that they fit Python naming conventions
waittimes_df = waittimes_df.rename(columns={'Code':'airport_code', 'Airport':'airport_name'})
```

```python
#drop unncessary columns and rows and set airport code as index
waittimes_df = waittimes_df.drop(columns=waittimes_df.columns[2:-1])
waittimes_df = waittimes_df.drop_duplicates()
waittimes_df = waittimes_df.set_index('airport_code')
waittimes_df
```
