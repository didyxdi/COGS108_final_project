```python
import pandas as pd
import numpy as np
```

```python
df_id = pd.read_csv('datasets/airports/airport_poneeded.csv')
df_id
```

```python
df_id = df_id[['Unnamed: 0', 'city_id']]
df_id
```

```python
df_po = pd.read_csv('datasets/city/uscities.csv')
df_po
```

```python
df_po = df_po[['city', 'population']]
df_po = df_po.reset_index()
df_po.columns = ['city_id', 'city_name', 'population']
df_po
```

```python
df_merged = pd.merge(df_id, df_po, on='city_id')
df_merged
```

```python
#Using Airport Code as Index and drop the column that we do not need anymore
df_merged = df_merged.drop(columns = ['city_id'])
df_merged.columns = ['airport_code', 'city_name', 'population']
df_merged = df_merged.set_index('airport_code')
df_merged
```

```python
df_merged.to_csv('population_cleaned.csv')
```
