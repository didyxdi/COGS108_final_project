```python
df_cities_need = pd.read_csv('cities_need.csv', index_col = "airport_code")
df_birdstrike_cleaned = pd.read_csv('birdstrike_cleaned.csv', index_col = "airport_code")
df_population_cleaned = pd.read_csv('population_cleaned.csv', index_col = "airport_code")
df_airport_poneeded = pd.read_csv('datasets/airports/airport_poneeded.csv', index_col = "airport_code")
df_commercial_service_enplanements = pd.read_excel('datasets/city/commercial_service_enplanements.xlsx')
#df_cities_need_bird = = pd.read_csv('cities_need.csv', index_col = "AirportCode")
```

```python
df_merged_X = pd.concat([df_cities_need, df_birdstrike_cleaned], axis = 1, sort = False)
df_merged_X#
```

```python
df_merged_X = pd.concat([df_merged_X, df_population_cleaned[["population"]]], axis = 1, sort = False)
df_merged_X#
```

```python
df_airport_poneeded = df_airport_poneeded[["latitude", "longitude", "altitude_ft", "city_id", "temp_avg", "pcp_avg"]]
df_merged_X = pd.concat([df_merged_X, df_airport_poneeded], axis = 1, sort = False)
df_merged_X = df_merged_X.dropna()
df_merged_X#
```

```python
df_commercial_service_enplanements = df_commercial_service_enplanements.rename(columns = {"Locid":"airport_code", "% Change":"enplanements_change", "CY 18 Enplanements":"enplanements_18", "CY 17 Enplanements":"enplanements_17"})
df_commercial_service_enplanements = df_commercial_service_enplanements.set_index("airport_code")
df_commercial_service_enplanements = df_commercial_service_enplanements[["enplanements_17", "enplanements_18", "enplanements_change"]]
```

```python
df_merged_X = pd.concat([df_merged_X, df_commercial_service_enplanements], axis = 1, sort = False)
df_merged_X = df_merged_X.dropna()
```
